```python
import numpy as np
from typing import List, Set, Optional

# The original (non-optimized) version
def create_bins_for_user_original(values: List[int], 
                                  n_bins: int, 
                                  exclude_id: Optional[int] = None) -> List[Set[int]]:
    indexed_values = tuple((idx, val) for idx, val in enumerate(values) if idx != exclude_id)

    min_val, max_val = min(indexed_values, key=lambda x: x[1])[1], max(indexed_values, key=lambda x: x[1])[1]
    bin_size = (max_val - min_val) / n_bins
    
    bins = [set() for _ in range(n_bins)]
    
    eps = 1e-6
    for idx, val in indexed_values:
        bin_idx = int((val - min_val) // (bin_size + eps))
        if bin_idx == n_bins:
            bin_idx -= 1
        bins[bin_idx].add(idx)
        
    return bins

# The optimized version
def create_bins_for_user_optimized(values: np.ndarray, 
                                   n_bins: int, 
                                   exclude_id: Optional[int] = None) -> List[Set[int]]:
    if exclude_id is not None:
        mask = np.arange(values.shape[0]) != exclude_id
        indexed_values = np.arange(values.shape[0])[mask]
        values = values[mask]
    else:
        indexed_values = np.arange(values.shape[0])

    min_val, max_val = np.min(values), np.max(values)
    bin_size = (max_val - min_val) / n_bins

    bins = [set() for _ in range(n_bins)]
    
    eps = 1e-6
    bin_indices = np.floor((values - min_val) / (bin_size + eps)).astype(int)
    bin_indices = np.clip(bin_indices, 0, n_bins - 1)
    
    for idx, bin_idx in zip(indexed_values, bin_indices):
        bins[bin_idx].add(idx)
        
    return bins

# Test cases
def test_equal_output():
    # Test case 1: Simple range of values
    values = [1, 2, 3, 4, 5, 6]
    n_bins = 3
    exclude_id = None
    
    result_original = create_bins_for_user_original(values, n_bins, exclude_id)
    result_optimized = create_bins_for_user_optimized(np.array(values), n_bins, exclude_id)
    assert result_original == result_optimized

def test_exclude_id():
    # Test case 2: Exclude an ID
    values = [1, 5, 10, 20]
    n_bins = 2
    exclude_id = 1  # Exclude the element with index 1 (value 5)
    
    result_original = create_bins_for_user_original(values, n_bins, exclude_id)
    result_optimized = create_bins_for_user_optimized(np.array(values), n_bins, exclude_id)
    assert result_original == result_optimized

def test_all_values_same():
    # Test case 3: All values are the same
    values = [10, 10, 10, 10]
    n_bins = 2
    exclude_id = None
    
    result_original = create_bins_for_user_original(values, n_bins, exclude_id)
    result_optimized = create_bins_for_user_optimized(np.array(values), n_bins, exclude_id)
    assert result_original == result_optimized

def test_large_range_of_values():
    # Test case 4: Large range of values
    values = [1, 10, 20, 100, 200, 500, 1000]
    n_bins = 5
    exclude_id = None
    
    result_original = create_bins_for_user_original(values, n_bins, exclude_id)
    result_optimized = create_bins_for_user_optimized(np.array(values), n_bins, exclude_id)
    assert result_original == result_optimized

def test_non_uniform_distribution():
    # Test case 5: Non-uniformly distributed values
    values = [1, 2, 2, 3, 100, 101]
    n_bins = 3
    exclude_id = None
    
    result_original = create_bins_for_user_original(values, n_bins, exclude_id)
    result_optimized = create_bins_for_user_optimized(np.array(values), n_bins, exclude_id)
    assert result_original == result_optimized

def test_exclude_id_out_of_bounds():
    # Test case 6: Exclude an ID that is out of bounds (should do nothing)
    values = [5, 15, 25, 35]
    n_bins = 2
    exclude_id = 10  # Non-existent ID
    
    result_original = create_bins_for_user_original(values, n_bins, exclude_id)
    result_optimized = create_bins_for_user_optimized(np.array(values), n_bins, exclude_id)
    assert result_original == result_optimized

# You can run these tests with pytest by simply running the command:
# pytest <name_of_test_file>.py
```

```python
test_equal_output()


test_exclude_id()


test_all_values_same()


test_large_range_of_values()


test_non_uniform_distribution()


test_exclude_id_out_of_bounds()

```
