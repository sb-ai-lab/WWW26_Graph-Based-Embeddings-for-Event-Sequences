```python
from ptls_extension_2024_research.latex_table_creation.latex_table_creation  import create_latex_table, get_metrics
from ptls_extension_2024_research.latex_table_creation.experiment_dicts_list_modifiers import sort_by_col, bolden_top_k
from ptls_extension_2024_research.latex_table_creation.prefix_map import prefix_map_from_idx_lst, get_idxs_where_all_metrics_superpass
```

```python
COLES_METRICS = {'AUC test': 0.877, 'ACC test': 0.793}
```

```python
with open("../../results/scenario_gender__coles_gnn__no_gnn_loss.txt") as f:
        file_content = f.read()
```

```python
get_metrics(
    file_content,
    "coles_gnn_weighted_2__GraphSAGE__no_coles_loss__no_orig____40_epoches",
    {
        "auroc": "auroc",
        "accuracy": "accuracy"
    }
)
```

```python

GNN_NAMES = ['GraphSAGE', 'GAT']
MAX_EPOCHES = 40
INCLUDE_GNN_USERS_IN_COLES_LOSS_STR = ['', 'gnn_users_in_coles_loss']
HAS_ORIG_STR = ['no_orig', 'has_orig']

MODEL_NAME = "CoLES + GNN (items). No gnn loss"
WEIGHTS_TYPE = "type 2"

experiment_dicts_list = []



def get_has_orig(experiment_name: str) -> str:
    if "has_orig" in experiment_name:
        return "yes"
    elif "no_orig" in experiment_name:
        return "no"
    else:
        raise ValueError(f"Experiments name doesn't contain information if embedding layer was used for item embeddings or not. Experiment name: {experiment_name}")


for gnn_name in GNN_NAMES:
    for gnn_users_in_coles_loss_str in INCLUDE_GNN_USERS_IN_COLES_LOSS_STR:
        for has_orig_str in HAS_ORIG_STR:
            experiment_name = f"coles_gnn_weighted_2__{gnn_name}__no_coles_loss__{has_orig_str}__{gnn_users_in_coles_loss_str}__{MAX_EPOCHES}_epoches"

            metrics = get_metrics(file_content, experiment_name, {'auroc': 'AUC test', 'accuracy': 'ACC test'})

            hyper_params = {
                "GNN": gnn_name,
                "Has item_id embedding": get_has_orig(experiment_name),
                "Has gnn users in coles loss": "yes" if gnn_users_in_coles_loss_str == 'gnn_users_in_coles_loss' else "no",
                "Train epochs": str(MAX_EPOCHES),
            }

            experiment_dicts_list.append({**hyper_params, **metrics})


experiment_dicts_list = sort_by_col(experiment_dicts_list, 'AUC test')
prefix_map = prefix_map_from_idx_lst(get_idxs_where_all_metrics_superpass(experiment_dicts_list, COLES_METRICS), "\n\\rowcolor{gray!50}\n")
experiment_dicts_list = bolden_top_k(experiment_dicts_list, 3, ['AUC test', 'ACC test'])
            


experiment_data = {
    MODEL_NAME: {
        WEIGHTS_TYPE: experiment_dicts_list
    }
}
```

```python
print(experiment_data)
```

```python
hyperparameter_header_strs_original = [
    r"\textbf{GNN}", r"\textbf{$\gamma$}", r"\textbf{Graph Task}", r"\textbf{$Loss_{wp}$}", r"\textbf{$\alpha$}", r"\textbf{Has item\_id embedding}", r"\textbf{Train epochs}", r"\textbf{AUC test}", r"\textbf{ACC test}",  
]

hyperparameter_header_strs = [
    r"\textbf{GNN}",  r"\textbf{Has item\_id embedding}", r"\textbf{Has gnn users in coles loss}", r"\textbf{Train epochs}", r"\textbf{AUC test}", r"\textbf{ACC test}",  
]


hyperparameters = ["GNN", "Has item_id embedding", "Has gnn users in coles loss", "Train epochs", "AUC test", "ACC test"]

caption = "Experiments where GNN was used as a feature aggregator and only coles loss was present."

row_prefix_dict = {
    MODEL_NAME: {
        WEIGHTS_TYPE: prefix_map
    }
}

latex_table = create_latex_table(experiment_data, hyperparameters, hyperparameter_header_strs, caption, row_prefix_dict)

print(latex_table)
```
