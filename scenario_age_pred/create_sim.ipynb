```python
import sys

import numpy as np
import torch
import dgl


# Paths
DATA_PATH = 'data/graphs/weighted/'
```

```python
def load_graph(graph_file_path):
    g_list, _ = dgl.load_graphs(graph_file_path, [0])
    g = g_list[0]
    return g

G = load_graph(DATA_PATH + 'train_graph.bin')
```

```python
G.num_nodes()
```

```python
n_users = 47000
n_items = 204

rows = np.arange(n_users)
cols = np.arange(n_items) + n_users

graph_emb_matr = G.adjacency_matrix().to_dense().numpy()
```

```python
graph_emb_matr.shape
```

```python
graph_adj_embs = graph_emb_matr[rows, :][:, cols]
```

```python
graph_adj_embs.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(graph_adj_embs)
sys.getsizeof(similarity_matrix)
```

```python
similarity_matrix.shape
```

```python
np.save("data/graphs/weighted/similarity_matrix.npy", similarity_matrix)
```
